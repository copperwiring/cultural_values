List all topics

```python
unique_topics = dollarstreet_data['topics'].unique()
print(len(unique_topics))
print(unique_topics)

# Print the number of rows for each topic
for topic in unique_topics:
    print(topic, dollarstreet_data[dollarstreet_data['topics'] == topic].shape[0])
```
